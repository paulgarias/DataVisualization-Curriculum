```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np

matplotlib.rcParams.update({'font.size':18,
                            'figure.facecolor': (1,1,1,1)})
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18)
```

```python
df = pd.read_csv("SightingsOffCoast.csv")
df.head(10)
```

```python
# rename column 
df = df.rename(columns={'Unnamed: 0':"Years"})
```

```python
df.head(10)
```

```python
df.plot(kind='line',x="Years",y=["Sharks","Dolphins","Whales"])
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df['Years'],df['Sharks'])
ax.set_xlabel("Years",fontdict={"size":22})
ax.set_ylabel("Number of sightings",fontdict={"size":22})
ax.legend(['Sharks'],loc='lower right')
ax.set_title("Sightings off of the coast")
```

```python
fig, ax = plt.subplots(figsize=(10,5))
df2 = df.set_index("Years")
ax.plot(df2)
ax.set_xlabel("Years",fontdict={"size":22})
ax.set_ylabel("Number of sightings",fontdict={"size":22})
ax.legend(df2.columns,loc='lower right')
ax.set_title("Sightings off of the coast",pad=30) # pad adds spacing between the title and the top of the axes (or plot area)
```
