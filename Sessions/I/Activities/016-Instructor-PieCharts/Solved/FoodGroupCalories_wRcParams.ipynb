```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np

matplotlib.rcParams.update({'font.size':16,
                            'figure.facecolor': (1,1,1,1)})
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18)
```

```python
df = pd.read_csv("FoodGroupCalories.csv")

# Display the first 5 entries in the DataFrame
df.head()
```

```python

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(aspect="equal"))

explode = (0,0,0,0.05)
wedges, texts, _ = ax.pie(df['Caloric Intake'].values, 
                         explode=explode, 
                         autopct='%1.1f%%',
                         shadow=False, 
                         startangle=0,
                         textprops={'color':'w',
                                   'fontweight':'bold'})

texts.clear()
ax.legend(wedges, df['Food groups'].values,
          title="Food groups",
          loc="center left", # make the center of the left edge of the legend box
          bbox_to_anchor=(1, 0.55)) # and place it at 100% position of the figures width, and 50% position of the figure's height
          # Think of bbox_to_anchor as an x,y coordinate of the loc

ax.set_title("Food group plot",fontdict={'fontsize':30})

plt.show()
```
