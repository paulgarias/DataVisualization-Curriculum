```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np

matplotlib.rcParams.update({'font.size':18,
                            'figure.facecolor': (1,1,1,1)})
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18)
```

```python
df = pd.read_csv("Cars.csv")
```

```python
df.head()
```

```python
df.plot(kind='scatter',x="SPEED",y="HP")
```

```python
fig, ax = plt.subplots(figsize=(10,5))
#Markers https://matplotlib.org/stable/api/markers_api.html
#Scatter documentation : https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter

ax.scatter(df.SPEED,df.HP,s=100,marker='s',edgecolors='black',c='red')
ax.set_xlabel("Speed (MPH)",fontdict={"size":22})
ax.set_ylabel("Max. Horsepower",fontdict={"size":22})

ax.set_title("Speed vs Max Horsepower",pad=30) # pad adds spacing between the title and the top of the axes (or plot area)
```

```python
model = LinearRegression()
```

```python
X = df.SPEED.values.reshape(-1,1)
y = df['HP'].values.reshape(-1,1)
```

```python
model.fit(X,y)
m,b = model.coef_, model.intercept_
ypred = model.predict(X) # produces a y vs x line that uses the points in df.SPEED
```

```python
Xminmax = [df.SPEED.min(),df.SPEED.max()]
ybound = m[0]*Xminmax + b[0] #makes a y vs x line that only uses min and max points
ybound
```

```python
fig, ax = plt.subplots(figsize=(10,5))
#Markers https://matplotlib.org/stable/api/markers_api.html
#Scatter documentation : https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter

ax.scatter(X,y,s=100,marker='s',edgecolors='black',c='red')
fig.text(0.2,0.5, f"y = {m[0][0]:3.2f}x + ({b[0]:3.2f})")
ax.plot(X,ypred,linewidth=2)
ax.set_xlabel("Speed (MPH)",fontdict={"size":22})
ax.set_ylabel("Max. Horsepower",fontdict={"size":22})

ax.set_title("Speed vs Max Horsepower",pad=30) # pad adds spacing between the title and the top of the axes (or plot area)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
#Markers https://matplotlib.org/stable/api/markers_api.html
#Scatter documentation : https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter

ax.scatter(X,y,s=100,marker='s',edgecolors='black',c='red')
fig.text(0.2,0.5, f"y = {m[0][0]:3.2f}x + ({b[0]:3.2f})")
ax.plot(Xminmax,ybound,linestyle='dashed',linewidth=2)
ax.set_xlabel("Speed (MPH)",fontdict={"size":22})
ax.set_ylabel("Max. Horsepower",fontdict={"size":22})

ax.set_title("Speed vs Max Horsepower",pad=30) # pad adds spacing between the title and the top of the axes (or plot area)
```
